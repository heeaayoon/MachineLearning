```python
penguin = pd.read_csv("data/penguins.csv")
penguin.info()
```

```python
# 결측치 버리기
penguin = penguin.dropna(axis=0)
```

```python
penguin.info()
```

```python
X = penguin[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]]
y = penguin['species']
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # trainX/testX/trainy/testy 로 나눔
```

```python
#모델 학습
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train) # fit => train 데이터셋으로 학습하라
```

```python
#예측 및 평가
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# 성능높이기
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) #fit에는 반드시 train 데이터만 넣기
X_test_scaled = scaler.transform(X_test)
```

```python
# 학습 다시
knn.fit(X_train_scaled, y_train)
```

```python
#다시 예측 및 평가
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# KNeighborsClassifier(n_neighbors=5) 라는 디폴트 값이 정말 최적인가? 
# n_neighbors 라는 매개변수의 값(k)을 변경시켜가면서 확인해보기

k_values = [1,3,5,7,9,11]
accuracies = []
for k in k_values:
    # 1.k에 따른 모델을 학습
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_scaled, y_train)
    # 2.예측
    y_pred_temp = knn_temp.predict(X_test_scaled)
    # 3.정확도 측정
    acc = accuracy_score(y_test, y_pred_temp)
    # 4.저장
    accuracies.append(acc)
    # 5.출력
    print(f"k={k} 정확도={acc:.4f}")

#최적의 k는?
best_k = k_values[np.argmax(accuracies)]
best_k
```

```python
# 과적합 ?
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
